## convert the NaN values from the cuisines and dish_liked

```python
cols = ['cuisines','dish_liked']
```

## Convert the dish_liked column as it is needed for the tfidf 

```python
def dish(x):
    x = x.lower()
    x = x.split(',')
    x = [i.strip() for i in x]
    return x
```

```python
df['dish_liked'] = df['dish_liked'].apply(dish)
```

## Convert for the cuisines columns

```python
def cusin(x):
    x = x.lower()
    x = x.split(',')
    x = [i.strip() for i in x]
    x = sorted(x)
    return ' '.join(x)
```

```python
df['cuisines'] = df['cuisines'].apply(cusin)
```

```python
import re

def clean_and_join_list(lst):
    # Join the list into a single string and remove unwanted characters
    joined = ''.join(lst)
    cleaned = re.sub(r"['\[\]]", "", joined).strip()
    cleaned = re.sub("[^a-zA-Z]"," ",cleaned)
    return cleaned
```

```python
df['dish_liked'] = df['dish_liked'].astype('object')
```

```python
df['dish_liked'] = df['dish_liked'].apply(clean_and_join_list)
```

```python
df.head()
```

```python
df['dish_liked'] = df['dish_liked'].apply(lambda x : x.split())
```

```python
from nltk.corpus import stopwords
import re
```

```python
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    # Remove unwanted characters using regex
    words = re.sub(r'[^a-zA-Z]', ' ', text)
    words = words.split()
    words = [i.strip() for i in words]
    filtered_text = ' '.join([word for word in words if word.lower() not in stop_words])
    return filtered_text
```

```python
df['dish_liked'] = df['dish_liked'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin, BaseEstimator

class DenseTransformer(TransformerMixin, BaseEstimator):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return X.toarray()
pipeline = Pipeline(
    steps=[
        ('tfidf', TfidfVectorizer()),
        ('to_dense', DenseTransformer()), 
        ('pca', PCA(n_components=1))
    ]
)
```

```python
sss = pipeline.fit_transform(df['dish_liked'])
```
